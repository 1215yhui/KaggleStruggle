```python
g = nx.Graph()
```

```python
g = nx.from_pandas_dataframe(data,source='Name',target='Type 1')
print(nx.info(g))
```

```python
plt.figure(figsize=(20, 20))
pos = nx.spring_layout(g, k=0.15)
nx.draw_networkx(g,pos,node_size=25, node_color='red')
plt.show()
```
